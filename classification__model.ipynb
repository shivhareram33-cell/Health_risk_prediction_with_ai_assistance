```python
# import required library

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score , confusion_matrix

import warnings
warnings.filterwarnings('ignore')
```

## first create model form doctor

```python
# load datase
df = pd.read_csv("./data2/doctor_data.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['health risk'] = df['health risk'].map({'High' :2 , 'Low' : 0 , "Medium" : 1})
```

saprate x and y

```python
x = df.drop(['health risk'] , axis=1)
y = df['health risk']
```

```python
# trin test split

x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.25 , random_state=9)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
y_train
```

```python
df['health risk'].value_counts()
```

## Randomforest - 

no. of decision tree used to predict and here no need of balanced data

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(class_weight='balanced')
forest.fit(x_train , y_train)
y_pred = forest.predict(x_test)
print(f"accuracy score by random forest is : {accuracy_score(y_test , y_pred)}")

# matric = confusion_matrix(y_test , y_pred)
# plt.figure(figsize=(7 , 5))
# sns.heatmap(matric , cmap='coolwarm' , annot=True)
# plt.show()
print(f"confusion metrics :")
print(confusion_matrix(y_test , y_pred))
```

## Balanced the data first by SMOTE

```python
# import sys
# !{sys.executable} -m pip install imblearn
```

```python
# from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import SMOTE

smote = SMOTE()

x_smote , y_smote  = smote.fit_resample(x_train , y_train)
print(x_smote.shape)
print(y_smote.shape)
```

## Gradiant boosting

```python
from sklearn.ensemble import GradientBoostingClassifier

grad = GradientBoostingClassifier()
grad.fit(x_smote , y_smote)
y_pred = grad.predict(x_test)

print(f"accuracy by gradient boosting is : {accuracy_score(y_test , y_pred)}")

print(f"confusion matrix is : ")
print(confusion_matrix(y_test , y_pred))
```

## Decision tree

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()
tree.fit(x_smote , y_smote)
y_pred = tree.predict(x_test)

print(f"accuracy by gradient boosting is : {accuracy_score(y_test , y_pred)}")

print(f"confusion matrix is : ")
print(confusion_matrix(y_test , y_pred))
```

In the medical dataset if patient actual in high risk but predict as low/medium that is create a problem 

here we can see RandomForest pediction low where it predicted high as low risk, hence we should go with random forest
